# (a) Derive the length of service rendered for each of the claim records.

```python
df['start_datetime'] = pd.to_datetime(df['Date of Service'].map(str) + ' ' + df['Service From (Hour)'].map(str) + ':' + df['Service From (Minute)'].map(str) + ':' + '00')
df
```

```python
#Get initial list of end_time
end_time = pd.to_timedelta(df['Service To (Hour)'].map(str) + ':' + df['Service To (Minute)'].map(str) + ':' + '00')
end_time
```

```python
#Find all rows where end_time equals '00:00:00' and convert it to a Timedelta of +1 days
diff = pd.to_timedelta((end_time == '00:00:00').astype(int), unit='d')
diff
```

```python
#Get initial list of end_datetime
end_datetime = pd.to_datetime(df['Date of Service'].map(str) + ' ' + df['Service To (Hour)'].map(str) + ':' + df['Service To (Minute)'].map(str) + ':' + '00')
```

```python
#Get final list of end_datetime
end_datetime += diff
df['end_datetime']=end_datetime 
df
```

```python
df['length_of_service'] = df['end_datetime'] - df['start_datetime']
df
```
